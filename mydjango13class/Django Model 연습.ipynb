## Comditions (OR)

```python
qs = Shop.objects.all()
qs.filter(pk__gte=5, pk__lt=10)  #AND       # (pk__gte=5, pk__lt=10)  -> 조건(conditions)
```

```python
from django.db.models import Q
```

```python
Q(pk__gte=5)
```

```python
Q(pk__gte=10)
```

```python
Q(pk__gte=5)&Q(pk__gte=10)
```

```python
Q(pk__gte=5)|Q(pk__gte=10)
```

```python
qs = Shop.objects.all()
qs.filter(
    Q(pk__gte=5)|Q(pk__gte=10)
)
```

```python
conditions = Q(pk__gte=5)
# conditions = conditions | Q(pk__lt=10)
conditions |= Q(pk__lt=10)

conditions
```

## 지정 범위의 Rows    1번 방식이 더 효율적!

```python
qs = Shop.objects.all()
qs = qs.filter(pk__lte=10)
qs.update(
    description = "not used description"
)
```

```python
qs = Shop.objects.all()
qs = qs.filter(pk__lte=10)
for shop in qs:
    shop.description = "not used description"
    shop.save()
```
